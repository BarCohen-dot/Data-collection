```python
df  = pd.read_csv("/content/#animalabuse.csv", on_bad_lines="skip") #נכתוב את הנתיב שלו כאן colab-לאחר שהעלנו את הקובץ ל
```

```python
df_cleaned_names = df.dropna(subset=['username']) #מקטגורית שמות null-שמירת הטבלה שבה נמחקו כול ערכיי ה
df_cleaned_names.info() # הדפסת המידע של הטבלה ה״נקייה״
```

```python
TwitterPerUser = df_cleaned_names.groupby('username').size() # ניקח את הטבלה הנקייה ונספור את כמות המופיעים של כול שם לאחר קיבוץ (שכמובן אם הופיע אז כמובן שהגיב)
TwitterPerUser
```

```python
TwitterPerUser.nlargest(5) # אפשר גם לפי -> top_users = TwitterPerUser.head(5)
```

```python
AfterDate2020 = df[df['date'] > '20-01-01'] # נוציא את התאריכים מהטבלה ה״רגילה״
AfterDate2020                               # נשים לב שכאן ה-20 הוא שנה
```

```python
df_cleaned_Comments = df.dropna(subset=['content']) #מקטגורית תגובות null-שמירת הטבלה שבה נמחקו כול ערכיי ה
                                                    #הדבר לא יתאפשר null אנחנו למעשה רוצים לספור את כמות התגובות שמכילות משהו מסויים, ועם
```

```python
df_contains_Dogs = df_cleaned_Comments[df_cleaned_Comments['content'].str.contains('Dog|dogֿֿ|Dogs|dogs')]
df_contains_Dogs
```

```python
df_contains_Cats = df_cleaned_Comments[df_cleaned_Comments['content'].str.contains('Cat|cat|Cats|cats')]
df_contains_Cats
```

```python
Union_6And8 = pd.concat([df_contains_Dogs, df_contains_Cats, AfterDate2020], join='inner') # לצורך מניעת כפילויות של עמודות join='inner' - נשתמש במפתח
Union_6And8                                                                                # איחוד של שלושת הטבלאות -> תאריכים, כלבים, חתולים
```
