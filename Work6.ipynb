```python
import requests

URL = "https://news.ycombinator.com/"
page = requests.get(URL)

print("Print by 'content' : \n--------------------\n",page.content) # אפשר לפי תוכן
print("=" * 1000)
print("Print by 'text' : \n-----------------\n",page.text) # אפשר לפי טקסט
```

```python
from bs4 import BeautifulSoup

soup = BeautifulSoup(page.content, "html.parser")
```

```python
title = soup.title.string # על מנת לבטל תגיות string
print(f"Title: {title}")
```

```python
a_tags = soup.find_all('a')
print(f"Number of `a` tags: {len (a_tags) }")
```

```python
pagespace_tag = soup.find(id="pagespace") # הראשון שמופיע במסמך id מציאת תגית ה
print(f"First tag with id 'pagespace': {pagespace_tag}")
```

```python
from collections import Counter                         #היא כלי לספירה אוטומטית של אובייקטים הנמצאים באוסף נתון ,Counter מחלקת

links = [a['href'] for a in a_tags                      # ( לצורך המרת הלולאה - [a['href'] ) רשימת הקישורים תהיה מכול הקישורים שנעבור עליהם בלולאה
         if 'href' in a.attrs and "https" in a['href']] #גם באתר עצמו וגם ברשימת הקישורים שאספנו href אנחנו בודקים אם לקישור יש את מאפיין

most_common_links = Counter(links).most_common(3)       # שימוש בסיפרייה ייעודית לצורכי ספירה
print("3 most frequent links containing 'https':")

for link, count in most_common_links:
    print(f"{link}: {count} times")
```
